```python
import numpy as np
arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])

a = arr[0:1, 0:2]
b = arr[0:1, 0:3]
c = arr[0,0:1]
d = arr[1,1:3]
e = arr[2,0:3]
f = arr[0:2,1:3]
g = arr[1:3,0:1]
h = arr[1:3,1:3]

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```
